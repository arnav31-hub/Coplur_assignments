```python
#1) Create a list of at least 15 numbers and Calculate the total sum of all integers in a list and find the arithmetic mean (average). Also Identify the largest and smallest numerical values within a provided list. And Count Even and Odd Numbers:

lst = [10,34,59,77,8,19,22,65,88,53,14,17,32,46,50]
print("This is the list:",lst)
sum = 0
for i in range(0,15):
    sum += lst[i]
print("sum of all numbers =",sum)

mean = sum/15
print("Arithmetic Mean=",mean)

mx = max(lst)
mn = min(lst)

e=0
o=0
for i in range(0,15):
    if lst[i]%2==0:
        e += 1
    else:
        o +=1
print("Even number =",e)
print("Odd number =",o)
```

```python
#4) Delete every instance of a specific value from a list. List: [5, 20, 15, 20, 25, 50, 20] Item to remove: 20:

lst = [5, 20, 15, 20, 25, 50, 20]
print(f"Before removing 20:{lst}")
for i in range(0,7):
    if 20 in lst:
        lst.remove(20)
print(f"After removing 20:{lst}")
```

```python
#5) Write a Python program to create a dictionary by mapping two equal-length lists, one containing keys and the other containing values. keys = ["name", "age", "city"] and values = ["ABC", 25, "Jaipur"]:

keys = ["name", "age", "city"]
values = ["ABC", 25, "Jaipur"]
person = {}
for i in range(0,3):
    person[keys[i]] = values[i]
print(person)
```

```python
#6) Write a Python program to retrieve a specific value from a dictionary that is nested inside another dictionary. person = {"name": "ABC", "address": {"city": "Jaipur"}:

person = {"name": "ABC", "address":{"city":"Jaipur"}}
print(person["address"]["city"])
```
