```python
#1. Write a Python function to find the maximum of three numbers.

def maximum(n1,n2,n3):
    if n1>n2 and n1>n3:
        print(f"{n1} is greater")
    elif n2>n1 and n2>n3:
        print(f"{n2} is greater")
    else:
        print(f"{n3} is greater")

maximum(55,66,23)
```

```python
#2. Write a Python function that takes a list and returns a new list with distinct elements from the first list:

def dist(l1):
    l=len(l1)
    l2=[]
    for i in range(l):
        if l1[i] in l2:
            pass
        else:
            l2.append(l1[i])
    return l2

l=[13,14,15,14,13,16,15]
print(f"Before distinct elements:{l}")
result=dist(l)
print(f"After distinct element:{result}")
```

```python
#3. Write a Python function to multiply all the numbers in a list:

def mult(l1):
    l=len(l1)
    result=1
    for i in range(l):
        result *=l1[i] 
    return result

l=[2,3,4,5,6]
result=mult(l)
print(f"Multiplication of all numbers in list:{result}")
```

```python
#4. Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument:

def fact(n):
    if n==1 or n==0:
        print(f"Factorial of {n} is 1")
    elif n < 0:
        print(f"Invalid number,{n} is negative integer")
    else:
        fact=1
        for i in range(2,n+1):
            print(i)
            fact*=i
        print(f"Factorial of {n} is:{fact}")
    
fact(5)
fact(-5)
```

```python
#5. Write a Python program to reverse a string:

def rev(s1):
    print(f"String before reversing:{s1}")
    s2 = s1[::-1]
    print(f"String after reversing:{s2}")

s =input("Enter a string to reverse:")
rev(s)
```

```python
#6. Write a Python function to check whether a number falls within a given range 10 and 50:

def between(n):
    if n>=10 and n<=50:
        print(f"{n} falls between 10 and 50.")
    else:
        print(f"{n} does not falls between 10 and 50")

between(40)
between(9)
```

```python
#7. Write a Python function to Print Even Numbers from a Given List:

def eve_or_odd(l1):
    l=len(l1)
    for i in range(l):
        if l1[i]%2==0:
            print(l1[i]," ")
        else:
            pass
    
l1 =[2,3,4,5,6,7]   
eve_or_odd(l1)
```

```python
#8. Write a Python function that takes a number as a parameter and checks whether the number is prime or not:

def prime(n):
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not a prime number")
            break
        else:
            print(f"{n} is a prime number")
            break

n = int(input("Enter a number"))
prime(n)
```

```python
#9. Write a Python function that accepts a string and counts the number of upper and lower case letters:

def coun(s):
    lower=0
    upper=0
    for i in s:
        if i.isupper()==True:
            upper+=1
        elif i==" ":
            pass
        else:
            lower+=1
    print(f"Lower case:{lower}")
    print(f"Upper case:{upper}")
s1="Python"
coun(s1)
```
